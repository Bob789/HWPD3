```python
df = df.drop('Area code', axis=1)
display(df)
```

```python
df['traveler'] = df['International plan'].apply(lambda x: 'tourist' if x == 'Yes' else 'homer')
display(df)
 
```

```python
def check_status_person(day, night):
    if pd.isna(day) or pd.isna(night):
        return 'unknown'
    if day > night:
        return 'day person'
    else:
        return 'night person'
df['talker_type'] = np.vectorize(check_status_person)(df['Total day minutes'], df['Total night minutes'])
display(df)
```
