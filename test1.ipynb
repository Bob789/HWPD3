```python
import numpy as np
from IPython.display import display
import pandas as pd
```

```python
#  df1
df1 = pd.DataFrame({
    'id': range(1, 11),
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Hana', 'Ilan', 'Jack'],
    'department': ['HR', 'IT', 'Finance', 'IT', 'HR', 'Finance', 'IT', 'HR', 'Finance', 'IT'],
    'salary': [6000, 7000, 7500, 7200, 5800, 8000, 6900, 6100, 8200, 7000],
    'bonus': [500, 600, np.nan, 550, 450, np.nan, 480, 470, 520, 510],
    'full_time': [True, True, False, True, False, True, True, True, False, True],
    'rating': [4.5, 3.8, 4.0, 4.2, 3.6, 4.9, 4.0, 3.9, 4.7, 4.1]
})
```

```python
#  df2
df2 = pd.DataFrame({
    'id': [1, 2, 3, 4, 5, 6, 11],
    'project': ['A', 'B', 'C', 'D', 'E', 'F', 'G'],
    'hours': [30, 45, 50, 40, 35, 60, 25],
    'status': ['complete', 'in progress', 'complete', 'in progress', 'complete', 'complete', 'not started'],
    'team_lead': ['Tom', 'Jerry', 'Tom', 'Anna', 'Anna', 'Tom', 'Jerry'],
    'location': ['NY', 'SF', 'TX', 'NY', 'SF', 'TX', 'LA'],
    'budget': [5000, 6000, 7000, 5500, 5300, 8000, 3000]
})
```

```python
display(df1)
display(df2)
```

```python
# merge
df3 = pd.merge(df1, df2, on='id', how='inner')
display(df3)
```

```python
# groupby
df4 = df3.groupby('department')['salary'].mean()
display(df4)
```

```python
# fillna
df5 = df3['bonus'].fillna(0)
display(df5)    
```

```python
# map
df6 = df3['full_time'].map({True: 'Yes', False: 'No' })
display(df6)
```

```python
df7 = pd.DataFrame()
df7['salary_after_bonus'] = df3['salary'] + df3['bonus'].fillna(0)
display(df7)    
```

```python
# filter
df8 = pd.DataFrame()
df8 = df1[df1['salary']  >= 7000][['salary']]

display(df8)
```

```python
#drop
df9 = pd.DataFrame()
df9 = df3.drop(columns = ['budget'], axis=1)
display(df9)
```

```python
#value counts
df10 = pd.DataFrame()
df10 = df3['department'].value_counts()
display(df10)
```
